```python
import numpy as np
import matplotlib.pyplot as plt
```

## Hacemos una tabla de tama√±o T y ubicamos T Claves 
### Queremos saber como se distribuyen

```python
T = 365
tabla = np.zeros(T)
cant_claves = T
for i in range(cant_claves):
    hash_ = np.random.randint(T)
    tabla[hash_] = tabla[hash_] + 1

plt.figure(figsize=(11,4))
plt.plot(tabla)
```

```python
intentos = []
for i in range(100):
    indice_random = np.random.randint(T)
    intentos.append(tabla[indice_random])

print("En el caso promedio, cada lista tenia longitud", np.mean(intentos))
```

## Hacemos una tabla de tama√±o T y ubicamos T Claves 
### Queremos saber como se distribuyen

```python
T = 128
tabla = np.zeros(T)
cant_claves = 1000
for i in range(cant_claves):
    hash_ = np.random.randint(10000) % T
    tabla[hash_] = tabla[hash_] + 1

plt.figure(figsize=(11,4))
plt.plot(tabla)
```
