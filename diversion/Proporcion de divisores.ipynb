```python
import matplotlib.pyplot as plt
from sympy import divisors
```

```python
def cantidad_de_divisores(n):
    return len(divisors(n))
    
def proporcion_de_divisores(n):
    return cantidad_de_divisores(n)/n
    
def promedio_hasta_n(n):
    total = 0
    for i in range(1, n+1):
        total += proporcion_de_divisores(i)
    total /= n
    return total
```

```python
plt.plot([proporcion_de_divisores(i) for i in range(2,1000)])
```

```python
plt.plot([promedio_hasta_n(i) for i in range(1,1000)])
```
